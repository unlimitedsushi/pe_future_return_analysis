```python
names_df = pd.read_csv('./sp500_companies.csv')
index = list(names_df['Symbol'].str.replace('.', '-'))
```

```python
ticker.earnings_history['epsActual']
```

```python
before_date = '2025-10-21'
after_date = '2026-01-21'

data = []
for x in index:
    ticker = yf.Ticker(x)
    price_before = ticker.history(start=before_date,end=pd.to_datetime(before_date) + pd.Timedelta(days=1), interval="1d")['Close'].values[0]
    price_after = ticker.history(start=after_date,end=pd.to_datetime(after_date) + pd.Timedelta(days=1), interval="1d")['Close'].values[0]
    try: 
        eps_df = ticker.quarterly_income_stmt.loc['Basic EPS']
    except Exception as e:
        print(f"Skipping {x}: {e}")
        continue
    past_eps = eps_df[eps_df.index <= before_date] #retrieves all the quarterly eps before the before_date
    eps = past_eps[past_eps.notna()].iloc[0]  #retrieve the most recent eps (that has no missing value)
    if np.isnan(eps_df[eps_df.index <= before_date].iloc[0]):
        print('nan', x)

    print(x, eps, price_before, price_after)
    data.append([price_before, price_after, eps])
    
```

```python
df = pd.DataFrame(data, index=index, columns=['price_before', 'price_after', 'eps'])
df
```
